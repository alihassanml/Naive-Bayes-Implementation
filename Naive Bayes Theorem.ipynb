### BernoulliNB
Why we use BernoulliNB ?
we use this algo when we have data in the form of yes/no OR 0/1 
so we use this 

@ e.g

hot  | cold  | today

yes  | no    | yes

### MultinomialNB

# ***Naive Bayes Theorem Practical Implementation***

```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

### So We have classfication problem 
1. over data is continues so we can use third veriation (3. GaussianNB)

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(X_train,y_train)
```

```python
y_pred = gaussian.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
corr = confusion_matrix(y_pred,y_test)
```

```python
sns.set_style('darkgrid')
sns.heatmap(corr,annot=True)
```
